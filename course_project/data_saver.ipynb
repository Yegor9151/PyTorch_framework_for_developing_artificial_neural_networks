```python
import matplotlib.pyplot as plt

import torch
from torchvision import io
```

```python
ROOT = '../data/leapGestRecog/train'

IDX = 1

SAVE = True
```

```python
class GetImgs:
    
    def __init__(self, imgs_from, imgs_to, name='img'):
        self.imgs_from = imgs_from
        self.imgs_to = imgs_to
        self.name = name
        
    def get_frame(self, save=False):
        
        self.frames = io.read_video(self.imgs_from, pts_unit='sec')[0]
        
        print(self.frames.shape)
        plt.imshow(self.frames[-1])
        
        if save:
            if type(self.frames) == torch.Tensor:
                self.frames = self.frames.numpy()

            for idx, frame in enumerate(self.frames):
                plt.imsave(f'{self.imgs_to}/{self.name}{idx}.png', frame)
```

```python
fist = GetImgs('../data/fist.mp4', f'{ROOT}/fist', f'fist{IDX}')
ok = GetImgs('../data/ok.mp4', f'{ROOT}/ok', f'ok{IDX}')
palm = GetImgs('../data/palm.mp4', f'{ROOT}/palm', f'palm{IDX}')
no_gest = GetImgs('../data/no_gest.mp4', f'{ROOT}/no_gest', f'no_gest{IDX}')
```

```python
no_gest.get_frame(SAVE)
```
